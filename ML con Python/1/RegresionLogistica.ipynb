### Lectura de fichero del Dataframe formateado con características y salida en Pandas

```python
# Función para el procesamiento del dataframe con características
def cargarDatasetTitanic(rutaCSV):

    # Carga del CSV
    df = pd.read_csv(rutaCSV)

    # Mostrar primeras filas por si quieres inspeccionarlo
    print("Primeras filas del dataset original:\n", df.head(), "\n")

    # Selección de características y salida 
    y = df['Survived']
    x = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']] 

    # Preprocesamiento
    x = pd.get_dummies(x, drop_first=False)
    dfFormateado = x.copy()
    dfFormateado['Survived'] = y

    return x, y, dfFormateado
```

```python
# Llamada a la función de carga del dataset formateado
x, y, df = cargarDatasetTitanic(rutaDataset)
```

### Visualización de distribución de los atributos con Matplotlib

```python
import matplotlib.pyplot as plt

# Como los datos han sido preprocesados anteriormente, hay que realizar las 
# gráficas de distribución los atributos en base a las nuevas filas

# Gráfca de edad
plt.hist(df["Age"].dropna(), bins=30, color="skyblue", edgecolor="black")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Distribución de Edad")
plt.show()
```

```python
# Gráfica de tarifa
plt.hist(df["Fare"], bins=30, color="green", edgecolor="black")
plt.xlabel("Tarifa")
plt.ylabel("Frecuencia")
plt.title("Distribución de Tarifa")
plt.show()
```
