## Ingresos mensuales

### Conjunto de datos sintéticos

```python
import numpy as np

np.random.seed(42)
x = 10 * np.random.rand(100, 1) 
y = 3 + 2.5 * x + np.random.randn(100, 1)

print("Tamaño del conjunto de datos:", len(x))
```

### Ajuste de un modelo de regresión lineal 

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Creación previa del dataframe para poder ajustar el modelo de regresión lineal
dataFrame = pd.DataFrame({
    'experiencia': x.flatten(),
    'ingresos': (y.flatten() * 1000).astype(int)  # Convertimos a dólares
})

regLin = LinearRegression()
regLin.fit(dataFrame[['experiencia']], dataFrame['ingresos'])

# Predicciones de extremos
xMinMax = pd.DataFrame({
    'experiencia': [dataFrame["experiencia"].min(), dataFrame["experiencia"].max()]
})
yPred = regLin.predict(xMinMax)

print("Extremo x:", xMinMax)
print("Extremo y:", yPred)
```

### Visualización de resultados

Gráfica regular

```python
import matplotlib.pyplot as plt
plt.plot(x, y, "b.")
plt.xlabel("Años de experiencia")
plt.ylabel("Ingresos mensuales (en miles de dólares)")
plt.title("Experiencia / Ingresos")
plt.show()
```

Gráfica con la línea de regresión

```python
plt.plot(xMinMax, yPred, "g-")
plt.plot(dataFrame['experiencia'], dataFrame['ingresos'], "b.")
plt.xlabel("Años de experiencia")
plt.ylabel("Ingresos mensuales (USD)")
plt.title("Regresión Lineal: Experiencia /  Ingresos")
plt.show()
```

## Científico ambiental

### Conjunto de datos sintéticos

```python
np.random.seed(42)
x = 1 + 19 * np.random.rand(100, 1)
a = 100
b = 10
ruido = np.random.randn(100, 1)
y = a / x + b + ruido

print("Tamaño del conjunto de datos:", len(x))
```

### Ajuste de un modelo de regresión lineal

```python
# Creación previa del dataframe para poder ajustar el modelo de regresión lineal
dataFrame = pd.DataFrame({
    'distancia': x.flatten(),
    'contaminacion': y.flatten()
})

regLin = LinearRegression()
regLin.fit(dataFrame[['distancia']], dataFrame['contaminacion'])

# Predicciones de extremos
xMinMax = pd.DataFrame({
    'distancia': [dataFrame["distancia"].min(), dataFrame["distancia"].max()]
})
yPred = regLin.predict(xMinMax)

print("Extremo x:", xMinMax)
print("Extremo y:", yPred)
```

### Visualización de los resultados

Gráfica regular

```python
plt.plot(x, y, "b.")
plt.xlabel("Distancia al centro (km)")
plt.ylabel("Concentración de PM 2.5 (μg/m³)")
plt.title("Distancia / Contaminación")
plt.show()
```

Gráfica con la línea de regresión

```python
plt.plot(xMinMax, yPred, "g-", label="Línea de regresión")
plt.plot(dataFrame['distancia'], dataFrame['contaminacion'], "b.", label="Datos reales")
plt.xlabel("Distancia al centro (km)")
plt.ylabel("Concentración de PM 2.5 (μg/m³)")
plt.title("Regresión Lineal: Distancia / Contaminación")
plt.show()
```
