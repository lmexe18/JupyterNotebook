```python
import random
import matplotlib.pyplot as plt

# Se crea una lista de notas aleatorias
notas = [random.randint(0, 100) for _ in range(10)]

# Se obntiene el máximo y el mínimo de dicha lista
def obtenerMaxMin(lista):
    resultado = max(lista), min(lista)
    return resultado

# Se obtiene el promedio de las notas
def obtenerPromedio(lista):
    media = sum(lista) / len(lista)
    return media

# Se obtienen dos listas, una de aprobados y otra de suspensos
# Se crea la variable aprobado para poder modificar la nota con la que se aprueba
def aprobadosYsuspensos(lista, aprobado):
    aprobados = sum(1 for nota in lista if nota >= aprobado)
    suspensos = len(lista) - aprobados
    return aprobados, suspensos

# Se guardan los resultados en variables
notaMax, notaMin = obtenerMaxMin(notas)
promedio = obtenerPromedio(notas)
aprobados, suspensos = aprobadosYsuspensos(notas, 50)

# Se muestran los resultados
print(f"Notas: {notas}")
print(f"Nota máxima: {notaMax}")
print(f"Nota mínima: {notaMin}")
print(f"Promedio: {promedio:.2f}")
print(f"Aprobados: {aprobados}")
print(f"Suspensos: {suspensos}")

# Se muestran los resultados con matplotlib
plt.figure(figsize=(10, 6))
plt.bar(range(1, 11), notas, color='#b266ff')
plt.xlabel('Estudiante')
plt.ylabel('Nota')
plt.title('Notas de los estudiantes')
plt.xticks(range(1, 11))
plt.ylim(0, 100)
plt.axhline(60, color='red', linestyle='--', label='Nota de corte')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
